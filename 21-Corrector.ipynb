```python
import base64

def image_to_base64(image_path: str) -> str:
    """
    Convierte un archivo de imagen a una cadena codificada en base64.
    
    Args:
        image_path (str): Ruta al archivo de imagen.
    
    Returns:
        str: Cadena en base64 de la imagen.
    """
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')
    
from typing import List, Dict

def get_images_base64(root_dir: str) -> List[Dict[str, str]]:
    """
    Recorre recursivamente un directorio en busca de archivos de imagen y devuelve una lista de diccionarios
    con el nombre de archivo y el contenido codificado en base64.

    Args:
        root_dir (str): Ruta al directorio raíz.

    Returns:
        List[Dict[str, str]]: Lista de diccionarios con las claves 'filename' y 'base64'.
    """
    image_extensions = {'.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff', '.webp'}
    
    images_base64 = []

    # Recorre todos los subdirectorios y archivos
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in sorted(filenames):
            ext = os.path.splitext(filename)[1].lower()
            if ext in image_extensions:
                img_path = os.path.join(dirpath, filename)
                base64_str = image_to_base64(img_path)
                images_base64.append({
                    "filename": filename,
                    "base64": base64_str
                })

    return images_base64

```

```python
from pydantic import BaseModel, Field
from typing import List, Optional, Literal

class Response(BaseModel):
    question: int
    response: Literal['A', 'B', 'C', 'D', 'No contestada']

class Exam(BaseModel):
    is_valid_exam: bool = Field(default=True, description="Indica si es un examen")
    city: str = Field(description="Nombre de la ciudad")
    initials: str = Field(description="Iniciales (son las 2 o 3 letras antes de _page2.pdf)")
    filename: str = Field(description="Nombre del fichero")
    responses: List[Response] = Field(description="Lista de respuestas del examen")
    #unknown_responses: List[Optional[int]] = Field(default=None, description="Lista de preguntas con respuestas desconocidas")
```

```python
image_data = images_list[13]

system_prompt = """
Eres un corrector de exámenes de tipo test.
El examen consta de 20 preguntas.
Cada pregunta tiene 4 posibles respuestas: A, B, C o D.
Tu tarea es analizar la imagen y elaborar la siguiente cadena de razonamiento (CoT):

### Paso 1. Identificar cómo marca el alumno sus respuestas.
En el examen el alumno habrá escrito marcas en las casillas de las respuestas.
Lo más habitual es que rodee la letra de la respuesta elegida con un círculo.
Algunos alumnos pueden marcar la respuesta otros símbolos como:
    - una cruz,
    - con un aspa,
    - con un tick,
    - una línea,
    - un cuadradro,
    - ...
Debes identificar cuál es la forma de marcar las respuestas del alumno.
Debes responder con la forma de marcar las respuestas del alumno, por ejemplo: 
    "en este examen la marca es un círculo"
No puedes responder en plural. La marca siempre debe ser singular

### Paso 2. Identificar las preguntas no contestadas del alumno.
Habrá preguntas que el alumno no ha contestado.
Una pregunta no contestada es aquella que no tiene ninguna marca en ninguna de las opciones (A, B, C o D).
Muestra las preguntas que no han sido contestadas por el alumno.
Ejemplo:
    Preguntas no contestadas:
    - Pregunta 1. No contestada
    - Pregunta 2. No contestada
    - Pregunta 5. No contestada

### Paso 3. Resto de preguntas
El resto de preguntas son las que el alumno ha contestado.
Indica las preguntas que el alumno ha contestado.
Ejemplo:
    Preguntas contestadas: 
    - Pregunta 1. Contestada
    - Pregunta 2. Contestada
    - Pregunta 5. Contestada     

### Paso 4. Preguntas marcadas.
De las preguntas contestadas, identifica marcas que se han hecho en las respuestas.
En la lista deben aparecer todos los números de las preguntas contestadas.
Ejemplo:
    Preguntas con marcas:
    - Pregunta 1. Respuesta A con círculo
    - Pregunta 3. Respuesta B con círculo y cruz. Respuesta D con círculo
    - Pregunta 5. Respuesta C con círculo
    - Pregunta 8. Respuesta A con círculo tachado

### Paso 5. Contestación a preguntas
De la lista de preguntas del paso 4, debes indicar la respuesta elegida por el alumno.
Los criterios son los siguientes:
- Si hay una respuesta marcada con la marca que se dijo en el paso 1, entonces la respuesta es la opción marcada.
- Sino, si hay una o varias respuestas marcadas y todas las marcas son diferentes a la que se dijo en el paso 1, entonces la respuesta es "Anulada".
- Sino, si hay varias respuestas marcadas con la marca del paso 1, entonces la respuesta es "Anulada". 
- Sino, si hay varias respuestas marcadas, pero sólo una de ellas tiene la marca del paso 1, entonces la respuesta es la opción marcada con esa marca.

### Paso 6. Contestación a cada pregunta (1 a 20)
Debes elaborar un resumen de las respuestas del examen juntando las respuestas de los pasos 3 y 5

### Paso Final. Resumen de respuestas.
Haz una lista ordenada de las respuestas del examen, desde la pregunta 1 a la 20.
Ejemplo:
    [A, B, No contestada, D, No contestada, B, C, D, Anulada, B, C, D, A, B, C, D, A, B, C, D]
"""

user_prompt = """
Analiza la imagen y responde a las siguientes preguntas:
1. ¿Cómo marca el alumno sus respuestas?
2. ¿Qué preguntas no han sido contestadas?
3. ¿Qué preguntas han sido contestadas?
4. ¿Qué marcas se han usado en las respuestas de las preguntas contestadas?
5. ¿Qué se ha contestado en cada pregunta del paso 4?
6. ¿Qué ha contestado el alumno en cada pregunta del examen (1 a 20)?
7. ¿Cuál es el resumen de respuestas del examen?
"""

response = google_client.chat.completions.create(
    model="gemini-2.0-flash",
    messages=[
            {"role": "system", "content": 
                [
                        {"type": "text", "text": system_prompt},
                        
                ]},
            {"role": "user", "content": 
                [
                        {"type": "text", "text": user_prompt},
                        {"type": "image_url",
                                 "image_url": {"url": f"data:image/jpeg;base64,{image_data['base64']}",
                                        "detail": "hight"}}
                ]}],
    temperature=0.0
)

print(response.choices[0].message.content)
```

```python
import json 

def extract_responses(image_data) -> Exam:   
    system_prompt = f"""
        Eres un asistente digital para la corrección de exámenes.
        Se te pasa el nombre de fichero que contiene:
         - Las iniciales del alumno (son las 2 o 3 letras antes de _page2.png)
         - La ciudad de realización del examen.
        
        Además, se te pasa una imagen que contiene las respuestas del alumno al examen.
        
        El examen consta de 20 preguntas.
        
        A cada pregunta se le asigna una de las siguientes respuestas posibles: "A", "B", "C", "D" o "No contestada".
        
        Casos en los que se debe dar "No contestada" como respuesta a una pregunta:
            1. No se ha marcado ninguna de las opciones (A, B, C o D).
            2. Se han marcado varias opciones (A, B, C o D).
            3. Se han marcado varias opciones pero se han tachado todas ellas.

        EN RESUMEN:
            - Para dar una respuesta A, B, C, o D a una pregunta la respuesta debe estar claramente marcada (normalmente con un círculo) y **no tachada**.
            - Se dará como "No contestada" una pregunta si se han tachado todas las respuestas que se han marcado.
            
        Casos en los que se debe dar una respuesta (A, B, C o D) a una pregunta:
            1. Se ha marcado una de las letras de respuesta (A, B, C o D).
            2. Aparecen varias letras de respuesta tachadas y una sola marcada y no tachada.

        El campo "is_valid_exam" se utiliza para saber si se ha podido procesar correctamente el examen.

        Un examen se considera válido si ocurre lo siguiente:
        1. Se han extraído correctamente las iniciales y la ciudad.
    """

    response = google_client.beta.chat.completions.parse(
        model=MODEL,
        messages=[
            {"role": "system", "content": 
                [
                        {"type": "text", "text": system_prompt},
                        
                ]},
            {"role": "user", "content": 
                [
                        {"type": "text", "text": "Procesa el nombre del fichero y extrae las iniciales y la ciudad"},
                        {"type": "text", "text": f"Nombre del fichero: {image_data['filename']}"},
                        {"type": "text", "text": "En la siguiente imagen están las respuestas que debes procesar"},
                        {"type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{image_data['base64']}",
                                        "detail": "hight"}}
                ]}],
        response_format=Exam,
        #temperature=0,
    )

    return json.loads(response.choices[0].message.content)
```
